# Import Libraries

```python
import os
import glob
import pandas as pd
```

# Find all the file names to merge

```python
data_folder = './data'
data_folder = './archive/stock_market_data/nasdaq/csv'

os.chdir(data_folder)
```

```python
# Not Used fileTypesToMerge; may use for daily close price or price in every minute data

fileTypesToMerge = '';
extension = 'csv';
all_filenames = [i for i in glob.glob('*' + fileTypesToMerge + '*.{}'.format(extension))]
all_filenames[:10]

# take a subset of the data for experiment
list_of_files_to_process = all_filenames #[:10]
list_of_files_to_process[:100]
```

```python
len(list_of_files_to_process)
```

# Insert Stock Ticker. Save data in new files

```python
from os.path import exists
row_total_count = 0
# modified = 'modified/'
modified = 'modified/'
for f in list_of_files_to_process:
    print(f);
    
    file_exists = exists(modified + f)
    if file_exists:
        # print('file exists');
        # break;
        continue;

    try:    
        df_s = pd.read_csv(f)
        df_s.insert(0, "Ticker", [f[:-4]] * df_s.shape[0], allow_duplicates = True)
        df_s.to_csv(modified + f);
        row_total_count += df_s.shape[0]
    except:
        continue;
    finally:
        continue;

row_total_count
```

# Combine all files in the list into one file

```python
# Combine

combined_csv = pd.concat([pd.read_csv(modified + f) for f in list_of_files_to_process]);

# export to csv
#combined_csv.to_csv( "../combined_data/stock-price-" + fileTypesToMerge + "-all-merged.csv", index = False, encoding = 'utf-8-sig')
combined_csv.to_csv( "../nasdq-stock-price-" + fileTypesToMerge + "-all-merged.csv", index = False, encoding = 'utf-8-sig')
```
