# Download stock price data to be used in the project for stock price prediction

```python
import yfinance as yf
```

```python
# Download pricing data for multiple stocks/tickers
# approach 1: put the tickers in a list and browse through the list and keep downloading data one ticker at a time
# approach 2: put the tickers in a list, and check if a library method is there to download for multiple tickers at once

download_period = "2y";
download_interval = "1d";
```

```python
# Stock Tickers to download data for
```

```python
tickers = [
'WMT',
'XOM',
'AAPL',
'UNH',
'MCK',
'CVS',
'AMZN',
'T',
'GM',
'F',
'ABC',
'CVX',
'CAH',
'COST',
'VZ',
'KR',
'GE',
'WBA',
'JPM',
'GOOGL',
'HD',
'BAC',
'WFC',
'BA',
'PSX',
'ANTM',
'MSFT',
'UNP',
'PCAR',
'DWDP']
```

```python
tickers[:5]
```

# Collect stock closing prices daily
# Download historical data using the daily time frame for the start and end date

```python
start = "2017-01-01";
end = "2019-12-31";
download_interval = "1d";
ticker_list = tickers;

# Download historical data using the daily time frame for the start and end date
data = yf.download(ticker_list, interval = download_interval, start = start, end = end )
closing_prices_daily = data['Close']
closing_prices_daily
```

```python
# closing_prices_daily_ticker_top = closing_prices_daily.T
closing_prices_daily.to_csv('./data/per-day-fortune-30-company-stock-price-data.csv');
```

```python
# Collect stock closing prices per minute
```

```python
download_interval = "1m";
download_period = "1w";
ticker_list = tickers;
data = yf.download(ticker_list, interval = download_interval, period = download_period)
closing_prices_minutely = data['Close']
closing_prices_minutely
```

```python
closing_prices_minutely.to_csv('./data/per-minute-fortune-30-company-stock-price-data.csv');
```
