```python
# Not Used fileTypesToMerge; may use for daily close price or price in every minute data

fileTypesToMerge = '';
extension = 'csv';
all_filenames = [i for i in glob.glob('*' + fileTypesToMerge + '*.{}'.format(extension))]
all_filenames_no_extension = [i[:-4] for i in glob.glob('*' + fileTypesToMerge + '*.{}'.format(extension))]
all_filenames[:10]

# take a subset of the data for experiment
list_of_files_to_process = list(set(all_filenames_no_extension).intersection(set(stocks_in_graph_data))) #[:10]
list_of_files_to_process[:100]
```

# Insert Stock Ticker. Save data in new files

```python
from os.path import exists
row_total_count = 0
# modified = 'modified/'
modified = 'modified/'
for f in list_of_files_to_process:
    f = f + ".csv";
    print(f);
    
    file_exists = exists(modified + f)
    if file_exists:
        # print('file exists');
        # break;
        continue;

    try:    
        df_s = pd.read_csv(f)
        df_s.insert(0, "Ticker", [f[:-4]] * df_s.shape[0], allow_duplicates = True)
        
        print(min(df_s['Date']))
        
        df_s.to_csv(modified + f);
        row_total_count += df_s.shape[0]
    except:
        continue;
    finally:
        continue;

row_total_count
```

# Combine all files in the list into one file

```python
# Combine

combined_csv = pd.concat([pd.read_csv(modified + f + ".csv") for f in list_of_files_to_process]);

# export to csv
#combined_csv.to_csv( "../combined_data/stock-price-" + fileTypesToMerge + "-all-merged.csv", index = False, encoding = 'utf-8-sig')
combined_csv.to_csv( "../nasdq-stock-price-" + fileTypesToMerge + "-all-merged.csv", index = False, encoding = 'utf-8-sig')
```

```python
# find dates of the stock prices
import datetime

min_date_to_work_with = datetime.date(1900,1,1);

for f in list_of_files_to_process:
    f = f + ".csv";
    
    file_exists = exists(modified + f)
    if file_exists:

        df_s = pd.read_csv(modified + f, parse_dates=['Date'], dayfirst=True)          
        #pd.to_datetime(df_s.day, dayfirst=True)
        #df_s["Date"] = df_s["Date"].astype('datetime64[ns]');
        
        min_ticker = min(df_s['Date'])
        #min_date = s.split("-");
        #min_ticker = datetime.date(int(min_date[2]), int(min_date[1]), int(min_date[0]));
        #print(min_ticker, min_date_to_work_with)
        if (pd.Timestamp(min_ticker) > pd.Timestamp(min_date_to_work_with)):
            min_date_to_work_with = min_ticker
            print(f, min_date_to_work_with);
```

```python
# find dates of the stock prices
import datetime

max_date_to_work_with = datetime.date(2025,1,1);

for f in list_of_files_to_process:
    f = f + ".csv";
    
    file_exists = exists(modified + f)
    if file_exists:

        df_s = pd.read_csv(modified + f, parse_dates=['Date'], dayfirst=True)          
        
        max_ticker = max(df_s['Date'])
        print(f, max_date_to_work_with);
        if (pd.Timestamp(max_ticker) < pd.Timestamp(max_date_to_work_with)):
            max_date_to_work_with = max_ticker
            print(f, max_date_to_work_with);
```
